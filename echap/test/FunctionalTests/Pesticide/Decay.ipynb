```python
import pandas
from alinea.pearl.pearl import *
from alinea.pearl.pearl_leaf import *
```

```python

def cumulative_sum(n):
    cum_sum = []
    y = 0
    for i in n:  
        y += i  
        cum_sum.append(y)
    return cum_sum


def tab_pearl(nb_step, dt):
    compound_parameters_epoxi = {'MolarMass':329.8, 'VapourPressure':0.00001, 'TemVapPre':20, 'WatSolubility':7.1, 'TemWatSol':20, 
    'ThicknessLay':0.0006,'DT50Pen':0.33,'DT50Tra':0.433, 'CofDifAir':0.428, 'TemDif':20, 'FacWas':0.0001, 'FraDepRex':0}
    compound_parameters_chloro = {'MolarMass':265.9, 'VapourPressure':0.0000762, 'TemVapPre':20, 'WatSolubility':0.81, 'TemWatSol':20, 
    'ThicknessLay':0.0006,'DT50Pen':0.14,'DT50Tra':0.23, 'CofDifAir':0.428, 'TemDif':20, 'FacWas':0.0001, 'FraDepRex':0}
    tab_t = [(1, 0, 0, 0, 0)]
    # Loop
    t = 0
    for i in range(nb_step):
        t += dt        
        p = PearlLeaf(compound_parameters_epoxi, global_climate = {'temperature_air':15})   
        tab_t.append(p.run(tab_t[i][0], dt, local_climate={'global_radiation':10,'temperature_air':20}))
    return tab_t


def dataframe_pearl(nb_step, dt):
    tab_t = tab_pearl(nb_step, dt)
    df_t = pandas.DataFrame(tab_t)
    df_t.columns = ['AmaCrp', 'AmaVolCrp', 'AmaWasCrp', 'AmaPenCrp', 'AmaTraCrp']
    df_t['time'] = range(0,nb_step*dt+1,dt)
    return df_t

```

```python
nb_step=25
dt=1

df_t = dataframe_pearl(nb_step, dt)
df_t['Vol'] = cumulative_sum(df_t['AmaVolCrp'])
df_t['Was'] = cumulative_sum(df_t['AmaWasCrp'])
df_t['Pen'] = cumulative_sum(df_t['AmaPenCrp'])
df_t['Tra'] = cumulative_sum(df_t['AmaTraCrp'])
```

```python
plot(df_t['time'], df_t['AmaCrp'], label='Surf')
plot(df_t['time'], df_t['Vol'], label='Vol')
plot(df_t['time'], df_t['Was'], label='Was')
plot(df_t['time'], df_t['Pen'], label='Pen')
plot(df_t['time'], df_t['Tra'], label='Tra')
legend( loc='upper right', numpoints = 1 )
show()
```

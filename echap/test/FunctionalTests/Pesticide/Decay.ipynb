## Model Overview

The objectives of the decay model are to simulate the effect of the architecture on the environmental loss, and to simulate the decay of doses acting on pests.


Two models have been chosen to fullfill these objectives : 

---

- The Pearl_leaf model (adapted from pearl), that simulates the physical processes responsible for the decay of surfacic doses of pesticide deposited on leaves. Pearl considered 4 processes responsible for that decay : 

    - the volatilisation (driven by wind and temperature)
    - the wash-off (driven by rain)
    - the penetration in the leaf (driven by time only)
    - the photodegradation (driven by light)
        
    
- The Milne model (driven by time only), that gives an estimate of the global decay of pesticide doses (within and on leaves), based on data related to the decay of biological efficacy of products.


- The echap decay model combines these two approaches by considering a surfacic decay as predicted by Pearl model, and a decay of penetrated doses given by Milne model. This choice results from the comparison of parameters that indicate a much faster decay rate on leaf surfaces (due to penetration) compared to the global decay of surfacic + penetrated dose. In other words, at the light of penetration rate used in Pearl, the global decay of the Milne model mostly relate to the decay of the penetrated fraction, the surfacic doses beeing very small well before efficacy drops down.


- The echap model gives the opportunity for new options for computing protectant and eradicant efficacy. In the Milne model, protectant efficacy (that is the ability to reduce germination) and the eradicant efficacy (that is the ability to slow down disease development) were both estimated using the total (surfacic + penetrated) doses. Here we propose to use the total dose for protectant efficacy, but only the penetrated dose for the eradicant efficacy.


- Technically, the echap model is build on the top of the pearl model and introducing a new compartment for inactive_penetrated dose. The flux from penetrated to inactive dose is taken as a first order kinetic (as in the Milne model). We also separate between surfacic and cuticular doses, to avoid overestimation of losses at the surface (the cuticular fraction is not sensitive anymore)

### Original Pearl and Milne, Global weather, Boigneville, second date

```python
weather = Boigneville_2010_2011()
seq = pandas.date_range(start="2011-05-11", periods=24*10, freq='H')
pe = PearlLeaf(pearl_parameters['Epoxiconazole'])
pc = PearlLeaf(pearl_parameters['Chlorothalonil'])
sim_pearl_epox = pe.daily_run(1,weather, seq[0], 10)
sim_pearl_chloro = pc.daily_run(1,weather, seq[0], 10)
m = MilneModel(milne_parameters['Epoxiconazole'])
sim_milne = m.decay(1, seq)
m = MilneModel(milne_parameters['Chlorothalonil'])
sim_milne_c = m.decay(1, seq)
```

```python
ax = sim_pearl_epox.plot()
sim_milne.rename(columns={'dose':'Milne_decay'},inplace=True)
sim_milne.plot(ax = ax)
ax = sim_pearl_chloro.plot()
sim_milne_c.rename(columns={'dose':'Milne_decay'},inplace=True)
sim_milne_c.plot(ax = ax)
```

- surfacic decay is much faster than Milne decay.Better consistency with milne to keep protectant effect linked to global dose and not surfacic dose only. 

- For epoxyconazole, almost no sensitivity to environmental losses

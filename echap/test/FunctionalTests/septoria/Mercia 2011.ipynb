# Analysis of septoria measurements on Mercia wheat from Grignon in 2011

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pylab import savefig
from alinea.echap.disease.septo_data_reader import *
from alinea.echap.disease.septo_data_treatment import *
data, weather, adel = data_reader(year = 2011, variety = 'Mercia', from_file= 'control',
                                  wetness_duration_min = 10., temp_min = 0., temp_max = 25., 
                                  hide_old_data = True, with_adel = True)
```

## Contingency

```python
# Count notations by leaf for septoria on green
tab = table_count_notations(data, weather, variable = 'septo_green')
save_table_to_latex(tab, 'count_notations/mercia_2011_ctrl_green.tex')
tab
```

## Statistical description for separate top leaves

```python
# Get average confidence interval
df_mean, df_low, df_high = get_data_of_interest(data = data, weather = weather, 
                                                xaxis = 'degree_days', variable = 'septo_green', 
                                                leaves = range(1,5),
                                                minimum_sample_size = 0,
                                                error_bars = True)
df_int = (df_high - df_low)
df_int.mean().mean()
```

```python
# Septoria on green with statistics by leaf
df_mean, df_high, df_low = plot_confidence_and_boxplot(data, weather, leaves = range(1,7), variable='septo_green', 
                                                       xaxis = 'age_leaf', title_suffix='_control', xlims = [0, 900])
savefig('figures/mercia_2011_ctrl/stat_green.png', bbox_inches='tight')
```

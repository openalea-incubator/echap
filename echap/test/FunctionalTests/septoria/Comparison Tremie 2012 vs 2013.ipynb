# Comparative analysis of septoria measurements on Tremie wheat from Grignon in 2012 vs. 2013

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pylab import savefig
from alinea.echap.disease.septo_data_reader import *
from alinea.echap.disease.septo_data_treatment import *
wetness_duration_min = 10.
temp_min = 0.
temp_max = 25.
data_12, weather_12, adel_12 = data_reader(year = 2012, variety = 'Tremie12', from_file = 'control',
                                  wetness_duration_min = wetness_duration_min, temp_min = temp_min, temp_max = temp_max)
data_13, weather_13, adel_13 = data_reader(year = 2013, variety = 'Tremie13', from_file = 'control',
                                  wetness_duration_min = wetness_duration_min, temp_min = temp_min, temp_max = temp_max)
```

## Plot statistical description by leaf

```python
warnings.filterwarnings('ignore')
compare_confidence_and_boxplot_2012_2013(data_12, data_13, weather_12, weather_13, xaxis = 'degree_days', leaves=range(1,6),
                                               fig_size=(15, 10), minimum_sample_size = 5., xlims = [1200, 2100])
```

```python
compare_confidence_and_boxplot_2012_2013(data_12, data_13, weather_12, weather_13, xaxis = 'age_leaf_lig', leaves=range(1,6),
                                               fig_size=(15, 10), minimum_sample_size = 5., xlims = [0, 800])
```

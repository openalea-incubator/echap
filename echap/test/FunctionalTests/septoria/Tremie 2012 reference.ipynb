# Analysis of septoria measurements on Tremie wheat from Grignon in 2012

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pylab import savefig
from alinea.echap.disease.septo_data_reader import *
from alinea.echap.disease.septo_data_treatment import *
data, weather, adel = data_reader(year = 2012, variety = 'Tremie12', from_file = 'reference', 
                                  green_on_green = True, wetness_duration_min = 10., temp_min = 0., temp_max = 25.)
```

## Contingency

```python
# Count notations by leaf for septoria on green
tab = table_count_notations(data, weather, variable = 'septo_green')
save_table_to_latex(tab, 'count_notations/tremie_2012_ref_green.tex')
tab
```

```python
# Count notations by leaf for severity
tab = table_count_notations(data, weather, variable = 'severity')
save_table_to_latex(tab, 'count_notations/tremie_2012_ref_sev.tex')
tab
```

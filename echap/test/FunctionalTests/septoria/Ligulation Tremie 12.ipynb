# Analysis of ligulation dates in Tremie 2012

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pylab import savefig
from alinea.echap.disease.septo_data_reader import *
from alinea.echap.disease.septo_data_treatment import *
start_date = "2011-10-21"
end_date = "2012-07-18"
# Read weather data
start, end = format_date(start_date, end_date)
weather = read_weather(start, end, wetness_duration_min = 10, temp_min = 0., temp_max = 25.)

df_dates_mean = get_leaf_dates_from_file(filename = 'TTleaf_Tremie12.csv', event = 'ligulation', fnl = None)
df_dates_12 = get_leaf_dates_from_file(filename = 'TTleaf_Tremie12.csv', event = 'ligulation', fnl = 12)
df_dates_13 = get_leaf_dates_from_file(filename = 'TTleaf_Tremie12.csv', event = 'ligulation', fnl = 13)
```

```python
# Show data from Christian
df_dates_mean.columns = ["Mean"]
(df_dates_12, df_dates_13) = map(lambda x: x.sort(), (df_dates_12, df_dates_13))
df_dates_12.columns = ["fnl_12"]
df_dates_13.columns = ["fnl_13"]
pd.concat([df_dates_12, df_dates_13, df_dates_mean], axis = 1)
```

```python
# Plot data from Christian
fig, ax = plt.subplots(figsize = (8,6))
ax.plot(df_dates_12.values, df_dates_12.index, 'gd ')
ax.plot(df_dates_13.values, df_dates_13.index, 'r* ')
ax.set_yticks(df_dates_13.index)
ax.invert_yaxis()
ax.grid()
leg = ax.legend(["FNL 12", "FNL 13"], loc=2)
```

```python
# Read leaf data from disease notations
df = pd.read_csv(shared_data(alinea.echap, 'disease_measurements/tremis_2012_ctrl_ligulation.csv'), parse_dates={'datetime':[1]}, sep=';', index_col=[0])
df['degree_days'] = [int(weather.data['degree_days'][t]) for t in df.index]
```

```python
df_dates_mean_ = pd.DataFrame(columns = ["Mean"])
for lf in set(df['num_leaf_lig']):
    dates = []
    for pl in set(df['plant']):
        if lf in df[df['plant'] == pl]['num_leaf_lig'].values:
            dates.append(df[(df['num_leaf_lig'] == lf) & (df['plant'] == pl)]['degree_days'].iloc[0])
    df_dates_mean_.loc[lf, 'Mean'] = np.mean(dates)
    
df_dates_12_ = pd.DataFrame(columns = ["FNL 12"])
df_12 = df[df['fnl']==12]
for lf in set(df_12['num_leaf_lig']):
    dates = []
    for pl in set(df_12['plant']):
        if lf in df_12[df_12['plant'] == pl]['num_leaf_lig'].values:
            dates.append(df_12[(df_12['num_leaf_lig'] == lf) & (df_12['plant'] == pl)]['degree_days'].iloc[0])
    df_dates_12_.loc[lf, 'FNL 12'] = np.mean(dates)
    
df_dates_13_ = pd.DataFrame(columns = ["FNL 13"])
df_13 = df[df['fnl']==13]
for lf in set(df_13['num_leaf_lig']):
    dates = []
    for pl in set(df_13['plant']):
        if lf in df_13[df_13['plant'] == pl]['num_leaf_lig'].values:
            dates.append(df_13[(df_13['num_leaf_lig'] == lf) & (df_13['plant'] == pl)]['degree_days'].iloc[0])
    df_dates_13_.loc[lf, 'FNL 13'] = np.mean(dates)
```

```python
pd.concat([df_dates_12_, df_dates_13_, df_dates_mean_], axis = 1)
```

```python
# Plot data from Christian
fig, ax = plt.subplots(figsize = (8,6))
ax.plot(df_dates_12.values, df_dates_12.index, 'gd ')
ax.plot(df_dates_12_.values, df_dates_12_.index, 'g* ')
#ax.plot(df_dates_13.values, df_dates_13.index, 'rd ')
#ax.plot(df_dates_13_.values, df_dates_13_.index, 'r* ')
ax.set_yticks(df_dates_13.index)
ax.invert_yaxis()
ax.grid()
#leg = ax.legend(["FNL 12", "FNL 13"], loc=2)
```

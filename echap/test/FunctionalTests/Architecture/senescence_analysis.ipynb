## Plotting functions

```python
def plot_comparison_senescence_by_fnl(data, weather, degree_days=True, leaves = range(1,4), from_top = True, 
                                      fig_size=(8,6), pointer=True, title_suffix='', minimum_sample_size=5):
    from itertools import product
    
    # Set figure
    fig, ax = plt.subplots(figsize=(8,6))   
    
    # Get data and draw plot
    df_group = group_by_fnl(data, from_top = from_top)
    linestyles = iter(['-', '--', ':'])
    leg = []
    for fnl, data_fnl in df_group.iteritems():
        df_mean = get_data_of_interest(data = data_fnl, weather = weather, variable = 'necro', 
                                   xaxis = 'degree_days', leaves = leaves, minimum_sample_size = minimum_sample_size)
        ax = set_color_cycle(ax, df_mean)
        points, ax = plot_df_mean(ax, df_mean, marker='d', markersize=8, linestyle=linestyles.next(), linewidth=2)
        leg += ['F%d fnl %d' % lf_fnl for lf_fnl in product(df_mean.columns, [fnl])]
        if pointer==True:
            set_pointer(fig = fig, points = points, df_mean = df_mean, degree_days = degree_days)
    
    # Customize plot
    ax.set_ylim([0, 100])
    ax.grid(alpha=0.5)
    ax = custom_ylabel(ax = ax, variable = 'Natural senescence')
    ax = custom_xlabel(ax, degree_days)
    ax.set_title(data_fnl['variety'][0]+'_'+str(data_fnl.index[0][0].year)+title_suffix, fontsize=18)

    # Set legend
    if len(leg) == len(df_group.keys()) * len(leaves):
        hs = np.reshape(ax.get_lines(), (len(df_group.keys()), len(leaves))).T
        hs = hs.reshape(1, len(leaves)*len(df_group.keys()))[0]
        leg = ['F%d fnl %d' % lf_fnl for lf_fnl in product(leaves, df_group.keys())]
        ax.legend(hs, leg, title='Leaf number', loc='best', fontsize=12)
    else:
        #ax.legend(handles = _get_legend_handles(), labels = leg, title='Leaf number', loc='best', fontsize=12)
        ax.legend(ax._get_legend_handles(), leg, title='Leaf number', loc='best', fontsize=12)
```

```python
def get_ssi(data, weather):
    df_group = group_by_fnl(data)
    df_ssi = pd.DataFrame()

    for fnl in df_group.iterkeys():
        df = get_mean(df_group[fnl], column='necro', by_leaf=True)

        # Fill empty values
        df.interpolate(inplace=True)
        df.fillna(value=0., inplace=True)

        # Leaf is senescent if upper leaf is senescent
        df_mask = df == 100.
        df_mask = df_mask.iloc[:,:6]
        df_mask.columns += 1
        df[df_mask] = 100.

        df_ssi[fnl] = df.sum(axis = 1)/100. + fnl - max(df.columns)

    df_ssi = add_index_ddays(df_ssi, weather)
    return df_ssi

def plot_ssi(data, weather, csv_name = None):
    """ e.g. : csv_name = 'figures/ssi/ssi_tremie_2013.csv' """
    df_ssi = get_ssi(data, weather)
    df_ssi.reset_index(level=0, drop = True, inplace = True)
    if csv_name is not None:
        df_ssi.to_csv(csv_name, sep=';')
    df_ssi.plot(marker = '*', markersize = 10, linestyle = '')
```

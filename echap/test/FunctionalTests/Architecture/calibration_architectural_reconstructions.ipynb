# Plot figures to calibrate architectural reconstructions of Mercia, Rht3 and Tremie wheat canopies

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 15, 10
from alinea.echap.architectural_reconstructions import *
HS_converter = HS_fit()
parameters = reconstruction_parameters()
```

## Haun Stage

```python
fits = HS_fit(reset=True) 
obs = archidb.validation_data()
archi_plot.haun_stage_plot(obs.haun_stage, fits)
```

```python
fits = GL_fit(**parameters) 
obs = archidb.green_leaves_aggregated(HS_fit())
# Compare varieties
archi_plot.green_leaves_plot_mean(obs, fits, HS_fit())
```

## Dimension

```python
axepfits = axepop_fits(**parameters)
hs = HS_fit()
fits = {k: pgen_ext.WheatDimensions(axepfits[k].mean_nff(), hs[k].dHS_nff(),1.2) for k in hs}
obs = archidb.validation_data()
# Blade length
archi_plot.dimension_plot(obs.dimensions, fit = fits, dimension = 'L_blade')
```

```python
# Blade width
archi_plot.dimension_plot(obs.dimensions, fit = fits, dimension = 'W_blade')
```

```python
# Sheath length
archi_plot.dimension_plot(obs.dimensions, fit = fits, dimension = 'L_sheath')
```

```python
# Internode length
archi_plot.dimension_plot(obs.dimensions, fit = fits, dimension = 'L_internode')
```

```python
# Insertion height of leaves
archi_plot.dimension_plot(obs.dimensions, fit = fits, dimension = 'H_col')
```

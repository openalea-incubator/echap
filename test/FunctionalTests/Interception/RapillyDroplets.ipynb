```python
def fill_nans(df, bins):
    def funx(x):
        indices = np.isnan(x.values.tolist())
        f = interp1d([0]+midpoints[~indices].tolist()+[100],[0] + x[~indices].tolist()+[100])
        return f(midpoints)
    midpoints = (bins[:-1] + bins[1:]) / 2
    midpoints[-1] = max(bins[:-1]) + 1
    return df.apply(funx, axis=0)
```

```python
def write_droplets_and_diameters():
    # Gather weather data for all years
    timings = []
    for start_year in range(1998, 2006):
        weather_file = 'meteo'+ str(start_year)[-2:] + '-' + str(start_year+1)[-2:] + '.txt'
        meteo_path = shared_data(alinea.septo3d, weather_file)
        weather = Weather(data_file=meteo_path)
        seq = pd.date_range(start = str(start_year)+"-10-01 01:00:00",
                                end = str(start_year+1)+"-07-01 01:00:00", 
                                freq='H')
                                
        every_rain = rain_filter(seq, weather)
        timings.append(IterWithDelays(*time_control(seq, every_rain, weather.data)))
    
    # Generate dataframes from weather data
    exposed_area=1.
    size = 100000
    bins = np.arange(0.2,2,0.2).tolist() + np.arange(2,6,1).tolist() + [100]
    labels,bins = pd.cut(bins,bins,right=False,retbins=True)
    df_count = pd.DataFrame(index=labels[:-1], columns=np.arange(1, 13))
    df_drop = pd.DataFrame(index=labels[:-1], columns=np.arange(1, 13))
    df_diam = pd.DataFrame(index=labels[:-1], columns=np.arange(1, 13))
    for dur in df_count.columns:
        for ind in df_count.index:
            seqs = [ctrl.value.rain.values for rain_timing in timings for ctrl in rain_timing if 
                    ctrl.value.rain.sum()>0.2 and ctrl.value.rain.count()==dur and 
                    df_count.index[np.searchsorted(bins,ctrl.value.rain.mean(), side='right')-1]==ind]
            if len(seqs)>0.:
                Ndroplets = 0.
                for i in range(dur):
                    # Get typical rain sequence for this duration
                    rain = np.mean([seq[i] for seq in seqs])
                    # Get number of droplets for this sequence
                    rd = RainDist(rain, rain_type="shower")
                    pop = (rd.dv_sample(int(size)))
                    ros = rd.ro
                    popd = PopDrops(ros, pop, duration = 3600.)
                    Ndroplets+=popd.emited_droplets(exposed_area)/1e7
                    # Get diameter of droplets for this sequence
                    sd = popd.splashing_drops()
                    diam = numpy.mean([d.diameter for d in sd])
                # Write in DataFrames
                df_count[dur][ind] = len(seqs)
                df_drop[dur][ind] = Ndroplets
                df_diam[dur][ind] = diam
                
    # Fill 'NaN' in DataFrames by interpolation
    df_drop = fill_nans(df_drop, bins)
    df_diam = fill_nans(df_diam, bins)
    
    # Save DataFrames
    df_count.to_csv('contingency_table.csv')
    df_drop.to_csv('droplet_emission.csv')
    df_diam.to_csv('droplet_diameter.csv')
```

```python
def get_bins(intervals):
    return np.unique(reduce(lambda x,y : x+ y,map(lambda x: map(float,re.split(',',re.sub('[\[\]()]','',x))),intervals)))

def look_up_table(df, bins, intensity, duration):
    bins = get_bins(df.index)
    return df.ix[np.searchsorted(bins,intensity, side='right')-1][min(duration, len(df.columns)-1)]
```

## Run simulation to get decay

- for time step : may be not too much detail is needed with pearl (a few microclim or global clim, large step)at application 1, as ntot cur = 5 and all s penetrated in 2/4 days. More interesting the 4/5 days after second application

- for output : do table/graph showing who is ntop_cur with time and in second setr of graph, follow decay on a leaf with ntop

```python
_, weather = sample_weather()
climates = weather.split_weather(24,seq[0],2)
```

```python
for i,wdata in enumerate(climates):
    update_pesticides(g,wdata)
    do_record(g,wdata,echap_recorder, header={'iter':i, 'TT':adel.canopy_age})
```

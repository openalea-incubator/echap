# Compare indicators of STB epidemics measured in Grignon on Tremie wheat in 2012 and 2013

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pylab import savefig
from alinea.echap.disease.septo_data_reader import *
from alinea.echap.disease.septo_data_treatment import *
data_12, weather_12 = data_reader(year = 2012, variety = 'Tremie12', from_file = 'control')
data_13, weather_13 = data_reader(year = 2013, variety = 'Tremie13', from_file = 'control')
```

```python
import mpld3
mpld3.disable_notebook()
fig, ax = plt.subplots(figsize = (16,12))
labels = []
df_mean = plot_by_leaf(data_12, weather_12, variable='severity', leaves = range(1,5), 
             pointer=False, xaxis = 'age_leaf_vs_flag_lig', linestyle = '-', 
             ax = ax, fig = fig, return_df_mean = True, xlims = [0, 800], title = '', error_bars = True)
labels += ['L '+str(lf)+': Tremie 2012' for lf in df_mean.columns]
df_mean = plot_by_leaf(data_13, weather_13, variable='severity', leaves = range(1,5), 
             pointer=False, xaxis = 'age_leaf_vs_flag_lig', linestyle = '--',
             ax = ax, fig = fig, return_df_mean = True, xlims = [0, 800], title = 'Tremie 2012 vs Tremie 2013 control',  error_bars = True)
labels += ['L '+str(lf)+': Tremie 2013' for lf in df_mean.columns]


color_list = [next(ax._get_lines.color_cycle) for col in range(7)]
leg = ax.legend(ax._get_legend_handles(), labels, loc='center left', bbox_to_anchor=(1, 0.5))
labels, handles = zip(*sorted(zip(labels, ax._get_legend_handles()), key=lambda t: color_list.index(t[1].get_c())))
leg = ax.legend(handles, labels, loc='center left', bbox_to_anchor=(1, 0.5))
savefig('figures/figures_rapport_ECHAP/tremie_2012_vs_2013_ctrl_sev.png', bbox_inches='tight')
```

## Indicator of STB emergence: Date/Leaf age at 5% of disease curve

```python
variable = 'severity'
xaxis = 'age_leaf_vs_flag_lig'
threshold = 3
fit_function_name = None
df_dates_12 = get_date_threshold(data_12, weather_12, variable, xaxis, threshold, fit_function_name)
df_dates_13 = get_date_threshold(data_13, weather_13, variable, xaxis, threshold, fit_function_name)
fig, ax = plt.subplots(1,1, figsize = (8,6))
plot_mean_thr_date_by_leaf(df_dates_12, ax = ax)
plot_mean_thr_date_by_leaf(df_dates_13, ax = ax, legend = ['2012', '2013'])
ax.invert_xaxis()
```

## Indicator of STB colonization: Severity at given leaf age

```python
variable = 'severity'
age = 300
fit_function_name = None
df_value_at_age_12 = get_value_at_age(data_12, weather_12, variable, age, fit_function_name)
df_value_at_age_13 = get_value_at_age(data_13, weather_13, variable, age, fit_function_name)
fig, ax = plt.subplots(1,1, figsize = (8,6))
plot_mean_value_at_age_by_leaf(df_value_at_age_12, ax = ax, age = age)
plot_mean_value_at_age_by_leaf(df_value_at_age_13, ax = ax, age = age, legend = ['2012', '2013'])
ax.invert_xaxis()
```

## Dates with infection risk

Infection risk appears when:

* wetness duration > 10h
* 0°C < air temperature < 30°C
* At least one dispersal event (rain > 0.2 mm/h) in previous 5 days

```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

def get_df_dates(data):
    df_dates = pd.DataFrame(index = set(data['num_leaf_ref']))
    df_dates['degree_days'] = map(lambda x: np.mean(data['date_ligulation_leaf'][data['num_leaf_ref']==x]), df_dates.index)
    df_dates.columns = ['date']
    return df_dates

fig, axs = plt.subplots(3,1)
ax1 = axs[0]
ax2 = axs[1]
ax3 = axs[2]
xlims = [1000,1400]

plot_septo_infection_risk(weather = weather_2011, leaf_dates = get_df_dates(data_2011), ax = ax1,
                          xlims = xlims, title = '2011', xlabel = False, ylabel = False)

plot_septo_infection_risk(weather = weather_2012, leaf_dates = get_df_dates(data_2012), ax = ax2,
                          xlims = xlims, title = '2012', xlabel = False, ylabel = False)

plot_septo_infection_risk(weather = weather_2013, leaf_dates = get_df_dates(data_2013), ax = ax3,
                          xlims = xlims, title = '2013', xlabel = False, ylabel = False)
plt.tight_layout()
```

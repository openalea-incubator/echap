## Whole season, whole plant comparison

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pylab import savefig
from alinea.echap.disease.septo_data_reader import *
from alinea.echap.disease.septo_data_treatment import *
wetness_duration_min = 10.
temp_min = 0.
temp_max = 25.
rht3_2011_winter = data_reader_2011('rht3_2011_winter.csv')
mercia_2011_winter = data_reader_2011('mercia_2011_winter.csv')
```

```python
data_mercia, weather, adel_mercia = data_reader(year = 2011, variety = 'Mercia', from_file = 'control',
                                  wetness_duration_min = wetness_duration_min, temp_min = temp_min, temp_max = temp_max, hide_old_data = True)
data_rht3, weather, adel_rht3 = data_reader(year = 2011, variety = 'Rht3', from_file = 'control',
                                  wetness_duration_min = wetness_duration_min, temp_min = temp_min, temp_max = temp_max, hide_old_data = True)
```

```python
data_m = pd.concat([mercia_2011_winter, data_mercia])
change_index(data_m, new_index = ['datetime', 'num_leaf_top'])
```

```python
data_m = pd.concat([mercia_2011_winter, data_mercia])
#leaves = [1, 2, 9, 10, 11, 12]
change_index(data_m, new_index = ['datetime', 'num_leaf_top'])
df = plot_grouped_leaves(data_m, weather, variable='septo_green', degree_days=True, fig_size=(8,6), 
                            title_suffix=' tnt', ylims = [0, 20], title = 'mercia 2011: F10-12 winter; F1-2 spring', 
                            return_descriptor = True)
```

```python
#data = pd.concat([rht3_2011_winter, data_rht3])
data_r = pd.concat([rht3_2011_winter, data_rht3[data_rht3['num_leaf_top']<=3]])
#leaves = [1, 2, 9, 10, 11, 12]

df = plot_grouped_leaves(data_r, weather, variable='septo_green', degree_days=True, fig_size=(8,6), 
                            title_suffix=' tnt', ylims = [0, 20], title = 'mercia 2011: F10-12 winter; F1-2 spring', 
                            return_descriptor = True, fixed_color = 'r')
```

```python
fig, ax = plt.subplots(figsize = (8,6))
df_r = plot_grouped_leaves(data_r, weather, variable='septo_green', degree_days=True, fig_size=(8,6), 
                            title_suffix=' tnt', ylims = [0, 20], title = '', 
                            return_descriptor = True, fixed_color = 'r', ax = ax)
df_r.to_csv('mean_severity_rht3_2011.csv', sep=';')
df_m = plot_grouped_leaves(data_m, weather, variable='septo_green', degree_days=True, fig_size=(8,6), 
                            title_suffix=' tnt', ylims = [0, 20], title = 'F10-12 winter; F1-2-3 spring', 
                            return_descriptor = True, fixed_color = 'g', ax = ax)
df_m.to_csv('mean_severity_mercia_2011.csv', sep=';')

labels = ['Mean', 'Confidence\n interval \n', 'Rht3', 'Mercia']
proxys = [plt.Line2D((0,1),(0,0), color='k', marker='o', linestyle='-'),
          plt.Rectangle((0,0), 0,0, facecolor='k', alpha = 0.3),
          plt.Rectangle((0,0), 0,0, facecolor='r', alpha = 0.3),
          plt.Rectangle((0,0), 0,0, facecolor='g', alpha = 0.3)]
ax.legend(proxys, labels, bbox_to_anchor=(1.02, 0.5), loc=2, borderaxespad=0.)
savefig('rht3_vs_mercia_2011.png', bbox_inches='tight')
```
